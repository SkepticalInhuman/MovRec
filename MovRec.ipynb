```python
df1 = df1.merge(df2,on='id')
df1.head(1)
```

```python
df1.shape
```

```python
C = df1['vote_average'].mean()
C
```

```python
m = df1['vote_count'].quantile(0.9)
m
```

```python
high_movies = df1.copy().loc[df1['vote_count'] >= m]
high_movies.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
high_movies['score'] = high_movies.apply(weighted_rating, axis=1)
high_movies.head()
```

```python
high_movies = high_movies.sort_values('score', ascending=False)
high_movies[['title_x', 'vote_count', 'vote_average', 'score']].head(10)
```

# Content Based Filtering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
df1['overview'] = df1['overview'].fillna('')
tfidf_matrix = tfidf.fit_transform(df1['overview'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df1.index, index=df1['title_x']).drop_duplicates()
```

```python
def get_rec(title, cosine_sim=cosine_sim):
  i = indices[title]
  sim_scores = list(enumerate(cosine_sim[i]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:11]
  mov_ind = [i[0] for i in sim_scores]
  return df1['title_x'].iloc[mov_ind]
```

```python
get_rec('The Avengers')
```

```python
from ast import literal_eval
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df1[feature] = df1[feature].apply(literal_eval)
```

```python
def get_director(x):
  for i in x:
    if i['job'] == 'Director':
      return i['name']
  return np.nan
```

```python
def get_list(x):
  if isinstance(x, list):
    names = [i['name'] for i in x]
    if len(names) > 3:
      names = names[:3]
    return names
  return []
def get_list2(x):
  if isinstance(x, list):
    names = [i['name'] for i in x]
    return names
  return []
```

```python
df1['director'] = df1['crew'].apply(get_director)
df1['cast'] = df1['cast'].apply(get_list)

features = ['keywords', 'genres']
for feature in features:
    df1[feature] = df1[feature].apply(get_list2)
```

```python
df1[['title_x', 'cast', 'director', 'keywords', 'genres']].head()
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['keywords', 'genres', 'director', 'cast']
for feature in features:
    df1[feature] = df1[feature].apply(clean_data)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df1['soup'] = df1.apply(create_soup, axis=1)
```

```python
df1['soup'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df1['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title'])
```

```python
get_rec('The Avengers', cosine_sim2)
```
